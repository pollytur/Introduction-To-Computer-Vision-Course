```python
import wget
plt.rcParams["figure.figsize"] = (10, 15) # (w, h)
```

```python
train_images_names = []
for i in range(1,11):
    for j in range(1,5):
        url = f'https://github.com/RABKKK/IrisRecognition/raw/master/Data/Train/{i}_{j}.jpg'
        filename = wget.download(url)
        train_images_names.append(filename)
```

```python
test_images_names = []
for i in range(1,11):
        url = f'https://github.com/RABKKK/IrisRecognition/raw/master/Data/Test/{i}.jpg'
        filename = wget.download(url)
        test_images_names.append(filename)
```

```python
actualClassLabel = [2,3,8,5,4,7,1,6,9,1]
```

```python
# copypasted from lab4
def read_and_resize(filename, grayscale = False, fx= 0.5, fy=0.5):
    if grayscale:
      img_result = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
    else:
      imgbgr = cv2.imread(filename, cv2.IMREAD_COLOR)
      # convert to rgb
      img_result = cv2.cvtColor(imgbgr, cv2.COLOR_BGR2RGB)
    # resize
    img_result = cv2.resize(img_result, None, fx=0.5, fy=0.5, interpolation = cv2.INTER_CUBIC)
    return img_result


def showInRow(list_of_images, titles = None, disable_ticks = False):
   
  count = len(list_of_images) 
  for idx in range(count):
    subplot = plt.subplot(1, count, idx+1)
    if titles is not None:
      subplot.set_title(titles[idx])
      
    img = list_of_images[idx]
    cmap = 'gray' if (len(img.shape) == 2 or img.shape[2] == 1) else None
    subplot.imshow(img, cmap=cmap)
    if disable_ticks:
      plt.xticks([]), plt.yticks([])
  plt.show()
```

```python
train_list_of_images=[]
print("Train Images")
for i in train_images_names:
    train_list_of_images.append(read_and_resize(i)) 
showInRow(train_list_of_images, disable_ticks=True)
```

```python
test_list_of_images=[]
print("Test Images")
for i in test_images_names:
    test_list_of_images.append(read_and_resize(i)) 
showInRow(test_list_of_images, disable_ticks=True)
```

## [Points 40] Preprocessing and Feature Extraction

```python
from numpy.linalg import norm
# modified from lab4
# and from https://stackoverflow.com/questions/6090399/get-hog-image-features-from-opencv-python
def hog(img, cell_size, bin_n):
    gx = cv2.Sobel(img, cv2.CV_64F, 1, 0)
    gy = cv2.Sobel(img, cv2.CV_64F, 0, 1)
    mag, ang = cv2.cartToPolar(gx, gy)
    
    bin = np.int32(bin_n*ang/(2*np.pi))

    bin_cells = []
    mag_cells = []

    cellx = celly = cell_size

    for i in range(0,int(img.shape[0]/celly)):
        for j in range(0,int(img.shape[1]/cellx)):
            bin_cells.append(bin[i*celly : i*celly+celly, j*cellx : j*cellx+cellx])
            mag_cells.append(mag[i*celly : i*celly+celly, j*cellx : j*cellx+cellx])   
    
    hists = [np.bincount(b.ravel(), m.ravel(), bin_n) for b, m in zip(bin_cells, mag_cells)]
    hist = np.hstack(hists)

    return hist
```

```python
# Write you code here!
def preprocessing(img_name):
    img = cv2.imread(img_name, 0)
    equ = cv2.equalizeHist(img)
    return equ
    
def features(img, cell_size=8, bins=16):
    
    return hog(img, cell_size=cell_size, bin_n=bins)

features2 = features(preprocessing(train_images_names[0]))

print("Feature vector", features2)
```

## [Points 20] Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
```

```python
train_X = [] 
train_Y = []
test_X = []

for i in train_images_names:
    train_X.append(features(preprocessing(i)))
    train_Y.append(int(i.split('_')[0]))

for i in test_images_names:
    test_X.append(features(preprocessing(i)))

test_Y = actualClassLabel
```

```python
clf = SVC(probability=True)
pipeline = make_pipeline(StandardScaler(), clf)
pipeline = pipeline.fit(train_X, train_Y)
```

## [Points 40] Calcuate the accuracy

```python
# Write you code here!

accuracy = pipeline.score(test_X, test_Y)
predictedLabel = pipeline.predict(test_X)

print("Actual Class", actualClassLabel)
print("Predicted Class", predictedLabel)
print("Accuracy", accuracy)
```
